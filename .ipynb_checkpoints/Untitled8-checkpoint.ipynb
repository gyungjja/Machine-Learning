```python
import os
import tarfile
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split

DOWNLOAD_ROOT = "https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data"
TRAIN_PATH = os.path.join("datasets","train")
TRAIN_URL = DOWNLOAD_ROOT + "datasets/train/train.csv.tgz"
TEST_PATH = os.path.join("datasets","test")
TEST_URL = DOWNLOAD_ROOT + "datasets/test/test.csv.tgz"
HOUSING_PATH = os.path.join("datasets","housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/sample_submission.csv.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
        
    tgz_path = os.path.join(housing_path, "sample_submission.csv.gz")
    if not os.path.isfile(tgz_path):
        urllib.request.urlretrieve(housing_url, tgz_path)      
        sample_submission.csv.tgz=tarfile.open(tgz_path)
        sample_submission.csv.tgz.extractall(path=housing_path)
        sample_submission.csv.tgz.close()
        
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "sample_submission.csv")
    return pd.read_csv(csv_path)

def fetch_train_data(train_url=TRAIN_URL, train_path=TRAIN_PATH):
    if not os.path.isdir(train_path):
        os.makedirs(train_path)
        
    tgz_path = os.path.join(train_path, "train.csv.gz")
    if not os.path.isfile(tgz_path):
        urllib.request.urlretrieve(train_url, tgz_path)      
        train.csv.tgz=tarfile.open(tgz_path)
        train.tgz.extractall(path=train_path)
        train.csv.tgz.close()
        
def fetch_test_data(test_url=TEST_URL, test_path=TEST_PATH):
    if not os.path.isdir(test_path):
        os.makedirs(test_path)
        
    tgz_path = os.path.join(test_path, "test.csv.gz")
    if not os.path.isfile(tgz_path):
        urllib.request.urlretrieve(test_url, tgz_path)      
        test.csv.tgz=tarfile.open(tgz_path)
        test.tgz.extractall(path=test_path)
        test.csv.tgz.close()
        
def load_train_data(train_path=TRAIN_PATH):
    csv_path = os.path.join(train_path, "train.csv")
    return pd.read_csv(csv_path)

def load_test_data(test_path=TEST_PATH):
    csv_path = os.path.join(test_path, "test.csv")
    return pd.read_csv(csv_path)

if __name__=="__main__":
    fetch_housing_data()
    fetch_train_data()
    fetch_test_data()
    housing = load_train_data()
    train = load_train_data()
    test = load_test_data()
    print(housing.head())
    print(train.head())
    print(test.head())
```
